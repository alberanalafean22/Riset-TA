# SPLIT DATA

Total data 1008 pasangan (Multiband dan mask)

Rasio Split:

*   Train :70%
*   Test : 15%
*   Valid : 15%



```python
import os
import shutil
import random

# Path asal data
multiband_src = '/content/drive/MyDrive/DATA BURNED AREA LANDSAT 9/MULTIBAND'
mask_src = '/content/drive/MyDrive/DATA BURNED AREA LANDSAT 9/MASK'

# Path tujuan base
base_out = '/content/drive/MyDrive/DATA BURNED AREA LANDSAT 9/SPLITDATA'

# Buat folder output split
for split in ['TRAIN', 'VAL', 'TEST']:
    os.makedirs(os.path.join(base_out, split, 'MULTIBAND'), exist_ok=True)
    os.makedirs(os.path.join(base_out, split, 'MASK'), exist_ok=True)

# Ambil semua file dari multiband
file_list = [f for f in os.listdir(multiband_src) if f.endswith('.tif')]
file_list.sort()  # agar urutan tetap

# Acak dengan seed untuk reproducibility
random.seed(42)
random.shuffle(file_list)

# Tentukan jumlah data
total = len(file_list)
train_count = 706 #60% dari 1008
val_count = 151
test_count = 151

# Split data
train_files = file_list[:train_count]
val_files = file_list[train_count:train_count + val_count]
test_files = file_list[train_count + val_count:]

# Fungsi salin file
def copy_files(file_list, split_name):
    for filename in file_list:
        # Path asal
        mb_src = os.path.join(multiband_src, filename)
        mask_filename = filename.replace('multiband', 'mask')  # jika pola nama sama, atau sesuaikan
        mk_src = os.path.join(mask_src, mask_filename)

        # Path tujuan
        mb_dst = os.path.join(base_out, split_name, 'MULTIBAND', filename)
        mk_dst = os.path.join(base_out, split_name, 'MASK', mask_filename)

        # Salin
        if os.path.exists(mb_src) and os.path.exists(mk_src):
            shutil.copy2(mb_src, mb_dst)
            shutil.copy2(mk_src, mk_dst)
        else:
            print(f"File tidak ditemukan: {filename} atau masknya")

# Salin file ke masing-masing folder
copy_files(train_files, 'TRAIN')
copy_files(val_files, 'VAL')
copy_files(test_files, 'TEST')

print("Selesai melakukan split dan penyimpanan data ke folder SPLITDATA.")
```

```python
import os

# Fungsi untuk menghitung jumlah file di masing-masing subfolder
def count_files(base_dir):
    for split in ['TRAIN', 'VAL', 'TEST']:
        mb_path = os.path.join(base_dir, split, 'MULTIBAND')
        mk_path = os.path.join(base_dir, split, 'MASK')

        mb_count = len([f for f in os.listdir(mb_path) if f.endswith('.tif')])
        mk_count = len([f for f in os.listdir(mk_path) if f.endswith('.tif')])

        print(f"[{split}] MULTIBAND: {mb_count} file")
        print(f"[{split}] MASK     : {mk_count} file")
        print("-" * 40)

# Jalankan fungsi
count_files(base_out)
```
